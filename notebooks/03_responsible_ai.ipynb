## 4. O problema do empate (limitação central)

O empate quase não tem assinatura estatística: acontece justamente quando os times estão equilibrados. Forçar o modelo a prever mais empates (peso de classe) sobe o recall do empate, mas derruba a acurácia geral. Não existe almoço grátis.

## 5. As probabilidades são confiáveis?

Um modelo de previsão de resultados vale mais pela probabilidade do que pelo 'chute' final. A curva de calibração mostra se, quando o modelo diz '30% de empate', o empate realmente acontece ~30% das vezes.

## 6. Conclusões e uso responsável

- O modelo bate o baseline ingênuo com folga, mas **não substitui** julgamento humano: ele ignora lesões, escalações, contexto de torneio e fatores fora dos dados.
- O empate é estruturalmente difícil — qualquer uso prático deve tratar a previsão de empate com cautela.
- Como apoio à decisão (risco/aposta), o valor está nas **probabilidades calibradas**, não no rótulo final.
- Vieses: seleções com mais histórico têm Elo mais estável; equipes novas ou com poucos jogos são previstas com mais incerteza.
